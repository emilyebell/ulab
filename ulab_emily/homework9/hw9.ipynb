# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 11)

def plot_functions(x):
    #plots f(x)= x^2 and g(x) = sqrt(x) on same plot and returns f(x) and g(x) values
    
    f_x = x**2
    g_x = np.sqrt(x)
    
    plt.figure(figsize=(10, 8))
    plt.plot(x, f_x, color='blue', linestyle='-', label='f(x) = x^2')
    plt.plot(x, g_x, color='green', linestyle='--', label='g(x) = sqrt(x)')
    
    plt.xlabel("x", fontsize=14)
    plt.ylabel("y", fontsize=14)
    
    plt.title("Plot of f(x) = x^2 and g(x) = sqrt(x)", fontsize=16)
    plt.legend()
    plt.show()
    
    return f_x.tolist(), g_x.tolist()

f_values, g_values = plot_functions(x)
print("f(x) =", f_values)
print("g(x) =", g_values)
```

```python
# Your code here

x = np.linspace(0, 20, 200) # change domain

def plot_functions(x):
    #plots f(x)= x^2 and g(x) = sqrt(x) on same plot and returns f(x) and g(x) values
    
    f_x = x**2
    g_x = np.sqrt(x)
    sum_fx_gx = f_x + g_x
    
    plt.figure(figsize=(10, 8))
    
    #add x/y limits
    plt.xlim(0, 10) 
    plt.ylim(0, 30)
   
    plt.plot(x, f_x, color='blue', linestyle='-', label='f(x) = x^2')
    plt.plot(x, g_x, color='green', linestyle='--', label='g(x) = sqrt(x)')
    
    #vertical/horizontial lines
    plt.axvline(x=5) 
    plt.axhline(y=25)
    
    #add legend in upper left
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    
    plt.xlabel("x", fontsize=14)
    plt.ylabel("y", fontsize=14)
    
    plt.title("Plot of f(x) = x^2 and g(x) = sqrt(x)", fontsize=16)
    plt.legend()
    plt.show()
    
    return f_x.tolist(), g_x.tolist(), sum_fx_gx.tolist()

f_values, g_values, sum_values  = plot_functions(x)
print("f(x) =", f_values)
print("g(x) =", g_values)
print("f(x) + g(x) =", sum_values)
```

```python
# Import your module and/or your code here

from PIL import Image

def plot_image_with_luminance(file_path):
    # Load image and print in two subplots: one with regular image and one with pseduocolor 
    
    # load image and make into array
    img = np.asarray(Image.open(file_path))
    print(repr(img))
    
    # plot regular image
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.xlabel("ground")
    plt.ylabel("toward the sky")
    plt.title("Regular Armadillo")
    plt.imshow(img)
    
    # plot pseudocolor image
    lum_img = img[:, :, 0]
    plt.subplot(1, 2, 2)
    plt.xlabel("ground")
    plt.ylabel("toward the sky")
    plt.title("Cursed Armadillo")
    plt.imshow(lum_img, cmap="hot")
    
    # show both plots
    plt.tight_layout()
    plt.show()

plot_image_with_luminance('armadillo.jpg')
```
